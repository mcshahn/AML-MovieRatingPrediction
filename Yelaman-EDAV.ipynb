```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
basics_df = pd.read_csv('Data/title-basics.tsv', sep='\t', low_memory=False)
basics_df
```

```python
basics_df.info()
```

```python
ratings_df = pd.read_csv('Data/title-ratings.tsv', sep='\t', low_memory=False)
ratings_df
```

```python
ratings_df.info()
```

```python
merged = pd.merge(basics_df, ratings_df, on ="tconst")
merged.head()   
```

```python
merged.describe(include= "all")
```

```python
# Making visuals of the features in merged dataframe
plt.figure(figsize=(12,8))
sns.displot(merged['averageRating'], bins=50)
plt.xlabel('Average Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Average Rating')
plt.show()
```

```python
merged['titleType'].value_counts()
```

```python
movies_df = merged[merged['titleType'] == 'movie']
movies_df.head()
```

```python
movies_df.info()
```

```python
#replace all \N with NaN
movies_df.replace('\\N', np.nan, inplace=True)
movies_df.isnull().sum()
```

```python
#drop end year column
movies_df.drop('endYear', axis=1, inplace=True)
movies_df.head()
```

```python
movies_df.drop('titleType', axis=1, inplace=True)
movies_df.head()
```

```python
# rename column StartYear to release year
movies_df.rename(columns={'startYear':'releaseYear'}, inplace=True)
```

```python
movies_df
```

```python
#bar chart of genres
cmap = plt.get_cmap('Pastel1')
colors = [cmap(i) for i in np.linspace(0, 1, 8)]

movies_df['genres'].value_counts().plot(kind='pie', subplots=True, autopct='%1.0f%%', figsize=(10, 10), colors=colors)
plt.title('Pie Chart - Number of Titles by IMDb genres')
```

```python
#give me rows where genre is reality tv or talk show
reality_tv = movies_df[movies_df['genres'] == 'Reality-TV']
reality_tv.head()
```

```python
talk_show = movies_df[movies_df['genres'] == 'Talk-Show']
talk_show.head()
```

```python
# create a list of all genres
genres_list = []
for genres in movies_df['genres']:
    if isinstance(genres, str):
        genres_list.extend(genres.split(','))

# get unique genres and their counts
unique_genres, counts = np.unique(genres_list, return_counts=True)

# create a dataframe with unique genres and their counts
genres_df = pd.DataFrame({'Genre': unique_genres, 'Count': counts})

# sort the dataframe by count in descending order
genres_df.sort_values('Count', ascending=False, inplace=True)

# display the dataframe
print(genres_df)
```

```python
genres_df.plot(x='Genre', y='Count', kind='bar', figsize=(12,8))
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Number of Titles by IMDb genres')
plt.show()
```
